##  ðŸ”¹ Loan Application Data Overview
1. This dataset captures loan applications and includes demographic, financial, and credit history details of applicants. 
2. It helps in analyzing loan approval trends, predicting approval chances, and identifying risk factors.  

## âœ… Key Features in the Dataset
Feature	               Description
1. Person Age:-	                Age of the applicant.
2. Person Gender:-	            Male or Female.
3. Person Education:-	        Education level: Associate, Bachelor, Doctorate, High School, Master.
4. Person Income:-	            Monthly income of the applicant.
5. Person Experience:-	        Years of experience (e.g., 2, 3, 5, 10).
6. Person Home Ownership:-	    Type of home ownership: MORTGAGE, OWN, RENT, OTHER.
7. Person Loan Amount:-	        The amount of loan applied for.
8. Loan Intent	Purpose of the loan:- DEBT CONSOLIDATION, EDUCATION, HOME IMPROVEMENT, MEDICAL, PERSONAL, VENTURE.
9. Loan Intent Rate:-	        Interest rate based on loan intent.
10. Loan Percentage Income:-	    Loan amount as a percentage of income.
11. CB Person Credit History Length:-	Credit history length in years.
12. Credit Score:-	            Applicant's credit score (higher is better).
13. Previous Loan Defaults:-	    Whether the applicant has previous loan defaults (Yes/No).
14. Loan Status:-	                0 (Rejected) or 1 (Approved).

# âœ… Use Cases of This Dataset
Loan Approval Prediction

1. Build a machine learning model to predict loan approval chances based on applicant details.
Credit Risk Analysis

2. Identify high-risk applicants (e.g., low credit score, high loan percentage of income).
Gender & Education-Based Insights

3. Analyzed how gender and education influence loan approvals.
Loan Intent Trends

4. Find which loan purposes (e.g., Education, Medical, Venture) have higher approval rates.
Income vs. Loan Amount Analysis

5. Checked if applicants with higher income get better loan approval rates.

# Ensemble Regularization Techniques
# Combining multiple regularization approaches

```python
# Bagging with different regularization parameters
def create_regularized_models():
    models = [
        RandomForestClassifier(
            n_estimators=100,
            max_depth=depth,
            min_samples_split=split,
            min_samples_leaf=leaf,
            max_features=features,
            random_state=42
        )
        for depth in [None, 10, 20,30]
        for split in [2, 5, 10]
        for leaf in [1, 2,4]
        for features in ['sqrt', 'log2']
    ]
    return models
```

```python
# Voting Classifier for ensemble regularization
from sklearn.ensemble import VotingClassifier
```

```python
# Create an ensemble of regularized models
ensemble_models = create_regularized_models()
voting_classifier = VotingClassifier(
    estimators=[(f'model_{i}', model) for i, model in enumerate(ensemble_models)],
    voting='soft'
)
```

```python
# Fit and evaluate voting classifier
voting_classifier.fit(X_train, y_train)
voting_score = voting_classifier.score(X_test, y_test)
print("\nEnsemble Voting Classifier Score:", voting_score)
```
