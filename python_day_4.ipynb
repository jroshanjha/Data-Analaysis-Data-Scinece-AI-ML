# Assignment Operator

An assignment operator in programming assigns a value or the result of an expression to a variable, with the most basic being the equals sign (=), but also including shorthand compound operators like +=, -=, *=, which perform an operation and assign the result in one step (e.g., x += 5 is x = x + 5)

```python
Z = 0
# Z incresed by 5
Z += 5 # Z =  Z + 5   =? 0 + 5 => 5
Z


# left side = right side value assign to the left side .
```

# Logical Operators

- Logical:
- Logical operations combine multiple conditions and return a boolean result.

# AND ( both are should be true )
- /* Logical AND Operators ( both conditions should be true )  */
1. True , True -> True
2. Fasle, True -> False
3. True , False -> False
4. False, Falsle -> False

```python
a = 10;
b = 5;
condition1 = (a > b);  # true
condition2 = (b < 20);  # true
result = (condition1 and condition2); # Logical AND   # &&
print(f"Logical AND result: {result }");
```

# OR ( atleast one condition true )
- /* Logical OR Operators (either condition should be true) */
1. True , True -> True
2. Fasle, True -> True
3. True , False -> True
4. False, Falsle -> False



```python
a = 10;
b = 5;
condition1 = (a > b);
condition2 = (b < 20);
result = (condition1 or condition2); # Logical OR
print(f"Logical OR result: {result }");
```

# XOR ( should be only one condition true )
1. True , True -> False
2. Fasle, True -> True
3. True , False -> True
4. False, Falsle -> False

```python
# XOR Operators

a = 10;
b = 5;
condition1 = (a > b);
condition2 = (b < 20);
result = (condition1 ^ condition2); # Logical XOR
print(f"Logical XOR result: {result }");
```

```python
a = 5  # Binary: 101
b = 3  # Binary: 011
result = a ^ b
print(result)
# Output: 6
# (Binary calculation: 101 ^ 011 = 110, which is 6 in decimal)
```
