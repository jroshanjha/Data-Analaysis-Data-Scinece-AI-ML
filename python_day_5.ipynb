```python
num1 = 5
num2 = 8
# Convert to binary strings (remove '0b' prefix)
binary1 = bin(num1)[2:] # '101'
binary2 = bin(num2)[2:] # '1001'

# Apply bitwise OR (result is in decimal, works on underlying binary)
result_decimal = num1 | num2 # 13

# Convert result to binary
result_binary = bin(result_decimal)[2:] # '1101'

print(f"{num1} ({binary1}) OR {num2} ({binary2}) = {result_decimal} ({result_binary})")
# Output: 5 (101) OR 9 (1001) = 13 (1101)
```

## Identity Operators (is, is not)

Identity operators are primarily used in the Python programming language to determine if two variables refer to the exact same object in memory (object identity), as opposed to just having the same value (equality).


```python
f = 5
g = 5

print("f is g:", f is g)

# this is a bat, cat
# she is tall girl
# He is paly boy.
# It is animval
```

```python
print("f is not g:", f is not g)
```

```python
print(id(f))
print(id(g))
```

## Membership Operators (in, not in)

Python has two membership operators, in and not in, which are used to test whether a value or variable is found within a sequence (such as a string, list, tuple, set, or dictionary). They return a Boolean value (True or False).

```python
my_list = [1, 2, 3, 4, 5]

print("Is 3 in list:", 3 in my_list)

print("Is 6 not in list:", 6 not in my_list)
```
