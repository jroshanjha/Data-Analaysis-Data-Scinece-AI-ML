# Number System

- Python has built-in support for the standard numeric types, including integers, floating-point numbers, and complex numbers. It also allows representation and conversion between different number systems like binary, octal, decimal, and hexadecimal.

```python
num = 45
num = 54.6555
num = 3 + 12j  # x = 3 ,y= 12
```

## Python Numeric Types
1. Python's primary number types are classes that store numeric values:
 - int (Integers): Represents whole numbers (positive, negative, or zero) of unlimited length. They do not have a fractional part.
 - float (Floating-point numbers): Represents real numbers with a decimal point. They are accurate up to approximately 15 decimal places and can be written in scientific notation (e.g., 1.23e+3).
 - complex (Complex numbers): Represents numbers in the form of x + yj, where x is the real part and y is the imaginary part.

2. Number Systems Representation
 - Python uses prefixes to represent integers in different number systems, which are all internally stored as integer objects:


| Number System | Base (Radix) | Prefix      | Digits Used | Example (Decimal 15) |
| ------------- | ------------ | ----------- | ----------- | -------------------- |
| Decimal       | 10           | None        | 0–9         | 15                   |
| Binary        | 2            | `0b` / `0B` | 0, 1        | `0b1111`             |
| Octal         | 8            | `0o` / `0O` | 0–7         | `0o17`               |
| Hexadecimal   | 16           | `0x` / `0X` | 0–9, A–F    | `0xF`                |

##✅ 1. Decimal (Base-10) – 🌍 Human-friendly
It's the system we use in everyday life.

Uses 10 digits: 0-9.

Easy for humans to read, but not ideal for computers.

```python
print(23) # this is decimal number
```

```python
print(101.45) # this is floating number
```

```python
print(101 + 1j) # this is complex number
```

## ✅ 2. Binary (Base-2) – 💻 Computer-friendly
Uses only 2 digits: 0 and 1.

Computers work using transistors (on = 1, off = 0).

All data (text, images, audio, etc.) is ultimately represented in binary

```python
x = 10
print(bin(x))  # convert into binary

# 2| 10 |0
# 2| 5  |1
# 2| 2  |0
#  | 1  |

# Bottom to top
#1010
```

## ✅ 3. Octal (Base-8) – 🧮 Compact representation
Uses digits: 0-7.

Easier shorthand for binary (1 octal digit = 3 binary digits).

Used in older systems, like UNIX file permissions (e.g., chmod 755).

```python
# 8|10|2   # 0,1,2,3,4,5,6,7
  # 1

# 12 => botton to top

print(oct(10))
```

### Octa to decimal

```python
print(int(0o12)) # 10
```

```python
# 8 | 196 | 4
# 8 | 24  | 0
#   | 3   |


# botton to top => 304
```

```python
print(int(0o304)) # 196
```

## ✅ 4. Hexadecimal (Base-16) – 🔢 Efficient for programmers
Uses digits: 0-9 and A-F (for 10–15).

1 hex digit = 4 binary digits → Great for reading and debugging memory/data.

Used in:

Color codes in web dev (#FF5733)

Memory addresses in low-level programming (0x1A3F)

Machine code / assembly (e.g., 0x90 = NOP instruction)

10 => A , 11 => B , 12 => C , 13 => D, 14= > E , 15 => F

```python
# 16 | 10 |

# 10 => A

print(hex(10))
```

# Octa to Binary and Hexa

```python
# 8|196| 4
# 8| 24 |0
#    3

# 196 => 304

# 2^5  , 2^4 , 2 ^3 , 2^2 , 2^1 , 2^0
# 32     16     8     4     2     1

# # step:- 1  1 Octa digit => 3 binary digit

# 3  => 011
# 0 =>  000
# 4 =>  100

print(bin(0o304))
```

```python
# step :- 2  4 digits binary =>

# 0 1100 1000

# 011000100 -> 0000 1100 0100

# 2^5  , 2^4 , 2 ^3 , 2^2 , 2^1 , 2^0
# 32     16     8     4     2     1

# 4 -> 0100
# 12 -> 1100 -> C
# 0 -> 0000

print(hex(0o304))
```
